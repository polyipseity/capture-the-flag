```python
import cv2
import pandas
```

```python
image = cv2.imread("aura.tif", cv2.IMREAD_UNCHANGED)
```

```python
for bits in range(8):
    for idx, clr in enumerate("bgr"):
        cv2.imwrite(f"{clr}@{bits + 1}.png", (image[:, :, idx] & 2 ** bits) * (256 / 2 ** bits))
```

```python
df = pandas.DataFrame(image[:,:,0])
df = df.map(lambda v: "{0:b}".format(v)[-6:])
df.describe()
```
